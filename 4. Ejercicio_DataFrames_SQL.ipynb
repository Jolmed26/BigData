# Ejercicio aplicado de DataFrames y Spark SQL

```python
import findspark
findspark.init()

import pandas as pd
import pyspark
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder\
        .master("local[*]")\
        .appName('PySpark_Df')\
        .getOrCreate()
```

```python
elfifas = spark.read.csv("WorldCupPlayers.csv", header=True, inferSchema=True)

## Visualiza los datos
elfifas.show()

```

```python
## ¿que tipo de datos contiene cada variable?
elfifas.dtypes
```

```python
## ¿Cuantos registros hay?
elfifas.count()
```

```python
## Obtén los principales estadísticos de Position
elfifas.describe('Position').show()
```

```python
## Slecciona y muestra los registros distintos de 'Player Name','Coach Name'
elfifas.select("Player Name", "Coach Name").distinct().show()
```

```python
elfifas.select("Position", "Event").show()
```

```python
## ¿Cuantos partidos con el ID de 1096 ha habido?
elfifas.filter(elfifas["MatchID"] == 1096).count()
```

```python
## Muestra los datos donde la posicion haya sido C y el evento sea G40
elfifas.filter((elfifas.Position == 'C') & (elfifas.Event == "G40'")).show()
```

```python
## Utiliza Spark SQL para mostras los registros donde el MatchID sea mayor o igual a 20
elfifas.createOrReplaceTempView("elfifas")
spark.sql("select * from elfifas where MatchID >= 20").show()
```
